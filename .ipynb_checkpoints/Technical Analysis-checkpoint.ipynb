```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import talib as ta
```

```python
from ta_suite import ta_suite
from ta_signals import ta_signals
from trading import act
```

```python
nifty = pd.read_csv(r"C:\Users\1987h\Downloads\Data science, machine learning and Quant Trading\algorithmic trading\GetZerodhaData\NIFTY 50.csv")
```

TECHNICAL INDICATOR STRATEGIES

- Slow stochastic   

STOCH(high, low, close[, fastk_period=?, slowk_period=?, slowk_matype=?, slowd_period=?, slowd_matype=?])

- Bollinger Bands

BBANDS(real[, timeperiod=5, nbdevup=2, nbdevdn=2, matype=0])

```python
tt = ta_suite(nifty)
# Testing SS and BBand Strategy
df = tt.combine()
df = ta_signals(df)
df = df.ss_dbb()
```

```python
# Testing RSI and BBand Strategy
df1 = tt.combine()
df1 = ta_signals(df1)
df1 = df1.rsi_bb()
```

```python
action = act(1000, "signal", df)
final_cash, ret, pnl = action.act()
print(final_cash)
print(len(pnl))
plt.plot(pnl, label="Stoch plus BBand")
plt.ylabel("PnL")
plt.legend()
plt.show()
```

```python
action = act(1000, "signal", df1)
final_cash, ret, pnl = action.act()
print(final_cash)
print(len(pnl))
plt.plot(pnl, label="RSI plus BBand")
plt.ylabel("PnL")
plt.legend()
plt.show()
```
