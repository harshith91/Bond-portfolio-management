```python
model.fit(X_tr, y_train, epochs=10, validation_split=0.1, batch_size=32, shuffle=False)
```

```python
model.evaluate(X_te,y_test, batch_size=32)
```

```python
y_pred = model.predict(X_te)
y_test[y_test==0] = -1
y_pred[y_pred>=0.50] = 1
y_pred[y_pred<0.50] = -1
accuracy_score(y_test, y_pred)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Nadam
def clf_model4(learning_rate):
    """
    tuning the learning rate for the optimizer
    """
    model = Sequential()
    model.add(LSTM(64, input_shape=(X_tr.shape[1],X_tr.shape[2]), activation='softsign',return_sequences=True))
    model.add(LSTM(32,activation='softsign'))
    model.add(Dense(1, activation='sigmoid'))
    optimizer = Nadam(learning_rate=learning_rate)
    model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
    return model

model = KerasClassifier(build_fn=clf_model4, epochs=20, batch_size=50)
learn_rate = [0.001, 0.01, 0.1, 0.2] 
param_grid = {'learning_rate':learn_rate}
random = RandomizedSearchCV(model, param_grid)
random.fit(X_tr, y_train)
```

```python
print(random.best_score_, random.best_params_)
means = random.cv_results_['mean_test_score']
parameters = random.cv_results_['params']
mn = 0
params = {}
for mean, parameter in zip(means, parameters):
    if mean>mn:
        mn = mean
        params = parameter
```

best learning rate:

0.5587021907170614 {'learning_rate': 0.001}

```python
# Use scikit-learn to grid search the weight initialization

def clf_model5(init_mode):
    model = Sequential()
    model.add(LSTM(64, input_shape=(X_tr.shape[1],X_tr.shape[2]),kernel_initializer=init_mode,return_sequences=True,activation='softsign'))
    model.add(LSTM(32,kernel_initializer=init_mode,activation='softsign'))
    model.add(Dense(1,kernel_initializer=init_mode,activation='sigmoid'))
    optimizer = Nadam(learning_rate = 0.001)
    model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
    return model

model = KerasClassifier(build_fn=clf_model5, epochs=20, batch_size=50)
init_mode = ['uniform', 'lecun_uniform', 'normal', 'zero', 'glorot_normal', 'glorot_uniform', 'he_normal', 'he_uniform']
param_grid = dict(init_mode=init_mode)
random = RandomizedSearchCV(model, param_grid, cv=3)
random.fit(X_tr, y_train)
```

```python
random.best_params_
```

best initializer

{'init_mode': 'normal'}

```python
#Tuning the dropout rate
def clf_model6(dropout_rate1, dropout_rate2):
    model = Sequential()
    model.add(LSTM(64, input_shape=(X_tr.shape[1],X_tr.shape[2]),kernel_initializer='normal',return_sequences=True,activation='softsign'))
    model.add(Dropout(dropout_rate1))
    model.add(LSTM(32,kernel_initializer='normal',activation='softsign'))
    model.add(Dropout(dropout_rate2))
    model.add(Dense(1,kernel_initializer='normal',activation='sigmoid'))
    optimizer=Nadam(learning_rate=0.001)
    model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
    return model

model = KerasClassifier(build_fn=clf_model6,epochs=20,batch_size=50)
dropout_rate1 = [0.0, 0.2, 0.4,0.5,0.6,0.8]
dropout_rate2 = [0.0, 0.2, 0.4,0.5,0.6,0.8]
param_grid = {'dropout_rate1':dropout_rate1,'dropout_rate2':dropout_rate2}
random = RandomizedSearchCV(model, param_grid, cv=3)
random.fit(X_tr, y_train)
```

```python
print(random.best_score_, random.best_params_)
means = random.cv_results_['mean_test_score']
parameters = random.cv_results_['params']
mn = 0
params = {}
for mean, parameter in zip(means, parameters):
    if mean>mn:
        mn = mean
        params = parameter
```

```python
#Tuning the number of neurons
def clf_model7(neuron1, neuron2):
    model = Sequential()
    model.add(LSTM(neuron1,input_shape=(X_tr.shape[1],X_tr.shape[2]),kernel_initializer='normal',activation='softsign',return_sequences=True))
    model.add(LSTM(neuron2,kernel_initializer='normal',activation='softsign'))
    model.add(Dropout(0.8))
    model.add(Dense(1,kernel_initializer='normal',activation='sigmoid'))
    optimizer=Nadam(learning_rate=0.001)
    model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
    return model

model = KerasClassifier(build_fn=clf_model7,epochs=20,batch_size=50)
param_grid={'neuron1':[16,32,64,128],'neuron2':[16,32,64,128]}
random=RandomizedSearchCV(model,param_grid,cv=3)
random.fit(X_tr,y_train)
```
